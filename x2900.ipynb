```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Plot

```python
# plot setting
import matplotlib as mpl
mpl.rc('axes', grid=True)
mpl.rc('xtick', top=True)
mpl.rc('ytick', right=True)
```

## $B^{+}\to D^{+}D^{-}K^{+}$

```python
f0 = np.loadtxt('./readData/B+toD+D-K+/data1.txt')
f0_low = np.loadtxt('./readData/B+toD+D-K+/data1-low.txt')
fig0, ax0 = plt.subplots(facecolor='w')
plt.minorticks_on()
ax0.set_xlabel(r'$ m(D^{-}K^{+}) $''[GeV]')
ax0.set_ylabel('Candidates / 20MeV')
ax0.errorbar(f0[:, 0], f0[:, 1], yerr=abs(f0_low-f0[:,1]), fmt='o', label='Exp');
```

## $B^{0}\to D^{-}D^{0}K^{+} $

```python
f1 = np.loadtxt('./readData/B0toD-D0K+/data3.txt')
f1_low = np.loadtxt('./readData/B0toD-D0K+/data3-low.txt')
fig1, ax1 = plt.subplots(facecolor='w')
plt.minorticks_on()
ax1.set_xlabel(r'$ m(D^{-}K^{+}) $''[GeV]')
ax1.set_ylabel('Events / 40MeV')

ax1.errorbar(f1[:, 0], f1[:, 1], yerr=abs(f1[:, 1]-f1_low), fmt='o', label='Exp');
```

## $ B^{0}\to D^{*-}D^{*0}K^{+} $

```python
f2 = np.loadtxt('./readData/B0toDstar-Dstar0K+/data4.txt', delimiter=',')
f2_low = np.loadtxt('./readData/B0toDstar-Dstar0K+/data4-high.txt')

fig2, ax2 = plt.subplots(facecolor='w')

ax2.errorbar(f2[:, 0], f2[:, 1], yerr=abs(f2[:, 1]-f2_low), fmt='o');
```
